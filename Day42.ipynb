```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.title('2-D Graph')
plt.savefig('G1.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*',linestyle='dashed')
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.title('2-D Graph')
plt.savefig('G1.png')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')

plt.subplot(2,2,3)
plt.plot(x,y,'r')
plt.subplot(2,2,4)
plt.plot(x,y,'g')
```
