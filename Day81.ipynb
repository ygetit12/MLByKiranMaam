```python
#from sklearn.preprocessing import OneHotEncoder
#onehot=OneHotEncoder()
#np.array(onehot.fit_transform(data['salary']))

onehot=pd.get_dummies(data['salary'])
onehot
```

```python
data1=data.join(onehot)
data1
```

```python
data1=data1.drop('salary',axis=1)
data1
```

```python
data1=data1.drop('Department',axis=1)
data1
```

```python
X=data1.drop(columns='left')
y=data1['left']#Dependent variable
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model1=model.fit(X_train,y_train)
```

```python
y_pred=model1.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cf=confusion_matrix(y_pred,y_test)
cf
```

```python
accuracy_score(y_pred,y_test)
```
